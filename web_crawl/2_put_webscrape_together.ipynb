```python
import pandas as pd
import os
import re

def files_to_dataframe(directory):
    all_data = pd.DataFrame([])
    for file in os.listdir(directory):
        try:
            df_read = pd.read_csv(directory+file, encoding="utf_8_sig")
            print(file, df_read.shape)
            # df_read.drop_duplicates(inplace=True)
            ##print(file, df_read.shape)
            all_data = pd.concat([all_data, df_read])
        except Exception as e:
            print(f"Error reading {file}: {e}")
    return all_data
```

```python
# Example usage:
directory_path = 'C:\\programming_projects\\ASU\\web_crawl\\web_data\\'
df = files_to_dataframe(directory_path)
print( df.shape)
df.drop_duplicates(subset = ['depth', 'title', 'topic', 'word_count', 'char_count', 'page_text'], inplace=True)
print('drop dups page_text:', df.shape)
df.head()
```

```python
def clean_text(text):
  """Keep only alphanumeric characters and basic punctuation."""
  text = re.sub(r"[^a-zA-Z0-9.!?\s]", ' ', str(text))
  text = re.sub(r'\s+', ' ', text).strip()
  return text

df['word_count'] = df['page_text'].apply(lambda x: len(str(x).split()))
df['char_count'] = df['page_text'].str.len()

df = df[df['word_count'] > 50]
print("word count", df.shape)

df = df.drop_duplicates(subset=['title', 'word_count', 'char_count', 'page_text'])
print('drop dups page_text:', df.shape)

# Apply the function to the DataFrame column
df['cleaned_text'] = df['page_text'].apply(clean_text)

# Remove all whitespace characters (including \r, \n, \t, \f, \v, and spaces)
df['cleaned_text'] = df['cleaned_text'].str.replace(r'\s+', ' ', regex=True).str.strip()

df = df.drop_duplicates(subset=['title', 'word_count', 'char_count', 'cleaned_text'])
print('drop dups cleaned_text:', df.shape)

df = df[(df['url'].isna() == False)]
print('URL can not be NA:', df.shape)

df['word_count'] = df['cleaned_text'].apply(lambda x: len(str(x).split()))
df['char_count'] = df['cleaned_text'].str.len()
df = df[df['word_count'] > 50]
print("word count", df.shape)

df['title'] = df['title'].apply(clean_text)
print(df.shape)
df = df.reset_index(drop = True)
```

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0

def lang_detect(text):
  text = re.sub(r'[^a-zA-Z0-9.,!?\s]', '', str(text))
  language_detected = detect(text)
  return language_detected


for i in range(0, len(df['langauage'])):
   if pd.isna(df.loc[i, 'langauage']):
      try:
         df.loc[i,'langauage'] = lang_detect(df['cleaned_text'].loc[i])
      except:
        print(i, '-', df['cleaned_text'].loc[i][0:20], '\n' )
```

```python
print(df.shape)

df_clean = df[df['langauage'] == 'en']
print("language", df_clean.shape)

df_clean.reset_index(drop=True, inplace=True)
```
